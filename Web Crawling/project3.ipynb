```python
import requests
from bs4 import BeautifulSoup
import os
```

```python
url = 'https://www.glamira.com/'
```

```python
def get_urls(url):
    if url.startswith('http'):
        r = requests.get(url)
        soup = BeautifulSoup(r.text, 'html.parser')
        a_tags = soup.find('nav', id='menu_mega').find_all('a', href=True)
        urls = [a['href'] for a in a_tags]
        urls = list(set(urls))
    return urls
```

```python
def get_img_links(url):
    if url.startswith('http'):
        r = requests.get(url)
        soup = BeautifulSoup(r.text, 'html.parser')
        img_tags = soup.find_all('img')
        if img_tags:
            img_links = [img['src'] for img in img_tags]
            if img_links: 
                img_links = list(set(img_links))
                return img_links
```

```python
def download_img(url, folername="images"):
    if not os.path.exists(folername):
        os.makedirs(folername)
    img_data = requests.get(url).content
    filename = os.path.join(folername, url.split("/")[-1])
    with open(filename, "wb") as handler:
        handler.write(img_data)
        print("Downloaded", filename)
```

```python
total = []
for url in get_urls(url):
    img_links = get_img_links(url)
    total.extend(img_links)
```

```python
for item in total:
   download_img(item) 
```
